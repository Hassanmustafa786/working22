```python
%%writefile streamlit_app.py

import streamlit as st

# Function to generate social media post
def generate_social_media_post(input_text, max_tokens=50):
    inputs = tokenizer(input_text, return_tensors="pt")
    tokens = model.generate(
        **inputs,
        max_new_tokens=max_tokens,
        temperature=0.75,
        top_p=0.95,
        do_sample=True,
    )
    generated_post = tokenizer.decode(tokens[0], skip_special_tokens=True)
    return generated_post

# Streamlit UI
st.title("Social Media Post Generator")

# Input prompt from the user
input_prompt = st.text_input("Enter the topic or prompt for your social media post:")

if st.button("Generate Post"):
    if input_prompt:
        generated_post = generate_social_media_post(input_prompt)
        st.subheader("Generated Social Media Post:")
        st.write(generated_post)
    else:
        st.warning("Please enter a topic or prompt for your social media post.")

st.sidebar.text("About")
st.sidebar.info("This is a simple Streamlit app for generating social media posts using a pre-trained language model.")
```
